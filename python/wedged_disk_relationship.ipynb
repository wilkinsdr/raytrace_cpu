```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

plt.rcParams['figure.dpi'] = 300
plt.rcParams['xtick.labelsize'] = 16  # Change 12 to your desired font size
plt.rcParams['ytick.labelsize'] = 16 

colors = ['#000000', '#E69F00', '#56B4E9', '#009E73', '#D55E00']

r_is = 2.321 #r_is

# Define the range of x values for the left and right halves
x_left = np.linspace(-10, -r_is, 500)
x_right = np.linspace(r_is, 10, 500)

# Define the functions for the lines with different slopes for left and right halves
def line_slope_positive_left(slope):
    return slope * (x_left + r_is)  # Adjust x values to start from 0

def line_slope_negative_left(slope):
    return -slope * (x_left + r_is)  # Adjust x values to start from 0

def line_slope_positive_right(slope):
    return slope * (x_right - r_is)  # Adjust x values to start from 0

def line_slope_negative_right(slope):
    return -slope * (x_right - r_is)  # Adjust x values to start from 0

# Plot lines with slopes 1, -1, 0.3, -0.3, 0.5, -0.5
plt.plot(x_left, line_slope_positive_left(1), color=colors[3])
plt.plot(x_left, line_slope_negative_left(1), color=colors[3])
plt.plot(x_left, line_slope_positive_left(0.1), color=colors[1])
plt.plot(x_left, line_slope_negative_left(0.1), color=colors[1])
plt.plot(x_left, line_slope_positive_left(0.5), color=colors[2])
plt.plot(x_left, line_slope_negative_left(0.5), color=colors[2])

plt.plot(x_right, line_slope_positive_right(1), color=colors[3])
plt.plot(x_right, line_slope_negative_right(1), color=colors[3])
plt.plot(x_right, line_slope_positive_right(0.1), color=colors[1])
plt.plot(x_right, line_slope_negative_right(0.1), color=colors[1])
plt.plot(x_right, line_slope_positive_right(0.5), color=colors[2])
plt.plot(x_right, line_slope_negative_right(0.5), color=colors[2])

# Shade the regions between the lines for left and right halves
plt.fill_between(x_left, line_slope_positive_left(1), line_slope_positive_left(0.5), color=colors[3], alpha=0.8)
plt.fill_between(x_left, line_slope_negative_left(1), line_slope_negative_left(0.5), color=colors[3], alpha=0.8)
plt.fill_between(x_left, line_slope_positive_left(0.1), line_slope_positive_left(0), color=colors[1], alpha=0.3)
plt.fill_between(x_left, line_slope_negative_left(0.1), line_slope_negative_left(0), color=colors[1], alpha=0.3)
plt.fill_between(x_left, line_slope_positive_left(0.5), line_slope_positive_left(0.1), color=colors[2], alpha=0.3)
plt.fill_between(x_left, line_slope_negative_left(0.5), line_slope_negative_left(0.1), color=colors[2], alpha=0.3)

plt.fill_between(x_right, line_slope_positive_right(1), line_slope_positive_right(0.5), color=colors[3], alpha=0.8)
plt.fill_between(x_right, line_slope_negative_right(1), line_slope_negative_right(0.5), color=colors[3], alpha=0.8)
plt.fill_between(x_right, line_slope_positive_right(0.1), line_slope_positive_right(0), color=colors[1], alpha=0.3)
plt.fill_between(x_right, line_slope_negative_right(0.1), line_slope_negative_right(0), color=colors[1], alpha=0.3)
plt.fill_between(x_right, line_slope_positive_right(0.5), line_slope_positive_right(0.1), color=colors[2], alpha=0.3)
plt.fill_between(x_right, line_slope_negative_right(0.5), line_slope_negative_right(0.1), color=colors[2], alpha=0.3)

# Plot the circle at the origin with radius 1.5 and fill it in black
circle = Circle((0, 0), 1.5, color='black', fill=True)
plt.gca().add_patch(circle)

# Plot a yellow star at the center of the circle
plt.plot(0, 5, marker='*', markersize=15, color='orange')

# Plot the circle
circle = plt.Circle((0, 5), (5+r_is)*np.sin(np.pi/4), color=colors[3], fill=False)
plt.gca().add_patch(circle)


# Set equal aspect ratio
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```
