```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_r_isco(a):
    A = 1 + (1 - a**2)**(1/3) * ((1 + a)**(1/3) + (1 - a)**(1/3))
    B = np.sqrt(3 * a**2 + A**2)
    return 3 + B - np.sqrt((3-A) * (3 + A + 2*B))
```

```python
eff = 0.06#0.35
edd = 0.3
r_isco = get_r_isco(0)

r_values = np.arange(200)
theta_values = np.linspace(0, np.pi/2, num=200)

z_vals = (3/2)*(1/eff)*(edd)*(1 - np.sqrt(r_isco/r_values))
```

```python
def plot_ellipse(a, b, center_x=0, center_y=0, c='black'):
    # Create an array of angles
    theta = np.linspace(0, 2*np.pi, 100)

    # Parametric equations for an ellipse with center (center_x, center_y)
    x = center_x + a * np.cos(theta)
    y = center_y + b * np.sin(theta)

    # Plot the ellipse
    plt.plot(x, y, color=c)
    plt.fill(x, y, color=c, alpha=0.3)
    plt.title("Elliptical Disk", fontsize=20)
    
    plt.plot([2*a+6, 200], [0, 0], color=c, linewidth=2)
    plt.plot([-2*a-6, -200], [0, 0], color=c, linewidth=2)
    #plt.ylim(-10, 20)
    plt.xlim((-200, 200))
    plt.ylim((-8, 15))
```
