```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_r_isco(a):
    A = 1 + (1 - a**2)**(1/3) * ((1 + a)**(1/3) + (1 - a)**(1/3))
    B = np.sqrt(3 * a**2 + A**2)
    return 3 + B - np.sqrt((3-A) * (3 + A + 2*B))
```

```python
get_r_isco(0)
```

```python
eff = 0.35 #0.35
edd = 0.3
r_isco = get_r_isco(0)

r_values = np.linspace(0, 200, num=2000)
#theta_values = np.linspace(0, np.pi/2, num=200)
```

```python
z_vals = (3/2)*(1/eff)*(edd)*(1 - np.sqrt(r_isco/(r_values)))
```

```python
z_vals
```

```python
plt.plot(r_values[r_values >= r_isco], z_vals[r_values >= r_isco])
plt.plot(r_values[r_values >= r_isco], -z_vals[r_values >= r_isco])

plt.plot(-r_values[r_values >= r_isco], z_vals[r_values >= r_isco])
plt.plot(-r_values[r_values >= r_isco], -z_vals[r_values >= r_isco])

plt.xlabel("r values")
plt.ylabel("z values")
plt.title("z values vs. r values")

# Show the plot
plt.show()
```
