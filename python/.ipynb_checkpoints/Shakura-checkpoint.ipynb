```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_r_isco(a):
    A = 1 + (1 - a**2)**(1/3) * ((1 + a)**(1/3) + (1 - a)**(1/3))
    B = np.sqrt(3 * a**2 + A**2)
    return 3 + B - np.sqrt((3-A) * (3 + A + 2*B))
```

```python
get_r_isco(0.9)
```

```python
eff = 0.35 #0.35
edd = 1.1
r_isco = get_r_isco(0.9)

r_values = np.arange(200)
theta_values = np.linspace(0, np.pi/2, num=200)
```

```python
z_vals = (3/2)*(1/eff)*(edd)*(1 - np.sqrt(r_isco/r_values))
```

```python
z_vals[199]
```

```python
z_vals[54]
```

```python
plt.plot(r_values[r_values >= r_isco], z_vals[r_values >= r_isco])
plt.plot(r_values[r_values >= r_isco], -z_vals[r_values >= r_isco])

plt.plot(-r_values[r_values >= r_isco], z_vals[r_values >= r_isco])
plt.plot(-r_values[r_values >= r_isco], -z_vals[r_values >= r_isco])

plt.xlabel("r values")
plt.ylabel("z values")
plt.title("z values vs. r values")

# plt.ylim(-5, 10)
# plt.xlim(-20, 60)
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib as mpl

# Set global font sizes for better visibility
mpl.rcParams['figure.dpi'] = 300
mpl.rcParams['font.size'] = 12  # Adjust font size as needed
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 12


shift=10
# Create shaded ellipses with appropriate colors and labels
ellipse1 = Ellipse(xy=(-54 - r_isco-shift, 0), width=54*2, height=3.737*2, edgecolor='orange', fc='orange', lw=2, alpha=0.5, label='Elliptical Disk')
plt.gca().add_patch(ellipse1)  # Add the first ellipse to the current axes
ellipse2 = Ellipse(xy=(54 + r_isco+shift, 0), width=54*2, height=3.737*2, edgecolor='orange', fc='orange', lw=2, alpha=0.5)
plt.gca().add_patch(ellipse2)  # Add the second ellipse to the current axes

# Use color-blind friendly colors and add labels for clarity
plt.plot(r_values[r_values >= r_isco]+shift, z_vals[r_values >= r_isco], color='blue', label='Shakura-Sunyaev Disk')
plt.plot(r_values[r_values >= r_isco]+shift, -z_vals[r_values >= r_isco], color='blue')
plt.plot(-r_values[r_values >= r_isco]-shift, z_vals[r_values >= r_isco], color='blue')
plt.plot(-r_values[r_values >= r_isco]-shift, -z_vals[r_values >= r_isco], color='blue')
#plt.xlabel("r'Radius ($\mathrm{r}_\mathrm{g}$)'")
#plt.ylabel("z")
plt.title("Shakura-Sunyaev Disk vs. Elliptical disk")

# Set axis limits to ensure everything is visible
plt.ylim(-5, 5)
plt.xlim(-60, 60)
plt.xticks([])
plt.yticks([])

# Add a legend to help identify plots
plt.legend(fontsize=10)

# Show the plot with higher DPI
plt.figure(dpi=300)  # Set DPI for better resolution; adjust as needed
plt.show()

```
